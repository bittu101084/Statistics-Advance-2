3. What is the significance level in hypothesis testing, and why is it important?

> he significance level (denoted as Î±) in hypothesis testing is a threshold used to decide whether to reject the null hypothesis. It represents the maximum probability of making a Type I error, which occurs when the null hypothesis is incorrectly rejected (a "false positive").

Why is it important?
Defines the Standard of Evidence:

The significance level sets the bar for how strong the evidence needs to be before rejecting the null hypothesis. Common values are 0.05 (5%) or 0.01 (1%).

Controls the Risk of Error:

By choosing a smaller significance level (e.g., 0.01), you reduce the chances of a false positive. However, this also makes it harder to detect a true effect.

Guides Decision-Making:

If the p-value (probability of observing the results under the null hypothesis) is less than or equal to the significance level, we reject the null hypothesis. If it's larger, we fail to reject it.

14. What is a confidence interval, and how is it used to interpret statistical results?

> A confidence interval (CI) is a range of values derived from sample data that is likely to contain the true population parameter (e.g., mean or proportion) with a certain level of confidence. It provides an estimate of the uncertainty associated with a sample statistic.

Key Components of a Confidence Interval:
Point Estimate: The sample statistic (e.g., sample mean) that serves as the center of the interval.

Margin of Error: Accounts for variability and uncertainty, determined by the standard error and the critical value from the statistical distribution (e.g., Z or T).

Formula: $$ \text{Confidence Interval} = \text{Point Estimate} \pm (\text{Critical Value} \times \text{Standard Error}) $$

Confidence Level: The probability that the interval contains the true parameter. Common levels are 95% or 99%.

Interpretation:
A 95% confidence interval means that if you repeated the sampling process 100 times, approximately 95 of those intervals would capture the true population parameter.

Important: The CI does not say thereâ€™s a 95% chance that the true value is in your specific intervalâ€”it reflects long-term reliability.

Usage in Statistical Results:
Estimation:

Provides a range for the true parameter rather than a single point estimate, which might be misleading.

Hypothesis Testing:

If a hypothesized value (e.g., the null hypothesis) falls outside the confidence interval, it suggests the null hypothesis may be rejected.

Example: If a CI for a mean difference is (2.5, 5.3) and the null hypothesis assumes no difference (value = 0), the null would be rejected because 0 is not within the interval.

15. What is the margin of error, and how does it affect the confidence interval?

> The margin of error is a measure of the uncertainty or variability in a statistical estimate. It quantifies the maximum expected difference between the sample statistic (e.g., mean or proportion) and the true population parameter. Essentially, it helps define the width of a confidence interval.

How Is Margin of Error Calculated?
The margin of error is derived from the critical value (based on the confidence level) and the standard error: $$ \text{Margin of Error} = \text{Critical Value} \times \text{Standard Error} $$

Critical Value: Represents the number of standard deviations corresponding to the confidence level (e.g., 1.96 for 95% in a Z-distribution).

Standard Error (SE): Accounts for the variability in the sample data and decreases with larger sample sizes.

How Does Margin of Error Affect the Confidence Interval?
The margin of error directly determines the range around the sample statistic where the true population parameter is likely to fall. A larger margin of error results in a wider confidence interval, indicating greater uncertainty, while a smaller margin of error leads to a narrower confidence interval, suggesting higher precision.

Factors Affecting Margin of Error:
Confidence Level: Higher confidence levels require larger critical values, increasing the margin of error (e.g., 99% vs. 95%).

Sample Size: Larger sample sizes reduce the standard error, decreasing the margin of error and narrowing the confidence interval.

Variability: Greater variability in the data increases the standard error, leading to a larger margin of error.

20. What is an ANOVA test, and what are its assumptions?

> An ANOVA (Analysis of Variance) test is a statistical method used to compare the means of three or more groups to determine if there are significant differences between them. It evaluates whether the variability between group means is greater than the variability within the groups, using the F-distribution.

Types of ANOVA Tests:
One-Way ANOVA:

Compares means across multiple groups based on one factor (independent variable).

Example: Testing whether students' test scores differ across different teaching methods.

Two-Way ANOVA:

Compares means across groups based on two factors and can also test for interactions between the factors.

Example: Evaluating how both teaching method and study time affect test scores.

Assumptions of ANOVA:
Independence:

Observations must be independent of each other (no overlaps or dependencies).

Normality:

The data within each group should be approximately normally distributed. ANOVA is robust to minor deviations from normality but can be affected by severe non-normality.

Homogeneity of Variance:

The variances among the groups should be roughly equal. This assumption can be tested using methods like Levene's Test.

Random Sampling:

The data should be collected through random sampling or random assignment to ensure unbiased results.

How It's Applied:
State Hypotheses:

Null Hypothesis (Hâ‚€): Group means are equal.

Alternative Hypothesis (Hâ‚): At least one group mean is different.

Compute F-Statistic:

Calculate the ratio of variance between groups to variance within groups.

Compare F-Statistic to Critical Value:

Using the degrees of freedom and significance level, determine whether to reject the null hypothesis.

Post-Hoc Tests:

If the null hypothesis is rejected, conduct post-hoc tests (e.g., Tukey's HSD) to identify which groups differ.

```python
# PRACTICAL QUESTIONS

#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
#  the results.

import math
from scipy.stats import norm

# Function to perform Z-test
def z_test(sample_mean, population_mean, population_std_dev, sample_size):
    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / (population_std_dev / math.sqrt(sample_size))

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    return z_score, p_value

# Input data
sample_mean = 105
population_mean = 100
population_std_dev = 15
sample_size = 30

# Perform the Z-test
z_score, p_value = z_test(sample_mean, population_mean, population_std_dev, sample_size)

# Print results
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")

# Interpret results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample mean and the population mean.")
```

```python
#3.  Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

import math
from scipy.stats import norm

# Function for one-sample Z-test
def one_sample_z_test(sample, population_mean, population_std_dev):
    # Calculate sample mean and sample size
    sample_mean = sum(sample) / len(sample)
    sample_size = len(sample)

    # Calculate Z-score
    z_score = (sample_mean - population_mean) / (population_std_dev / math.sqrt(sample_size))

    # Calculate p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    return z_score, p_value

# Simulating sample data
# Example: Generate a sample of size 30 with mean=105 (replace with your own sample data)
sample = [105, 100, 110, 108, 95, 98, 104, 99, 102, 106, 107, 103, 101, 100, 109, 110, 108, 103, 104, 95,
          97, 98, 100, 106, 105, 101, 108, 99, 100, 104]

# Parameters
population_mean = 100  # Known population mean
population_std_dev = 15  # Known population standard deviation

# Perform Z-test
z_score, p_value = one_sample_z_test(sample, population_mean, population_std_dev)

# Output results
print(f"Sample Mean: {sum(sample)/len(sample):.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")

# Interpret results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample mean and the population mean.")
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to perform a two-tailed Z-test
def two_tailed_z_test(sample_mean, population_mean, population_std_dev, sample_size, significance_level):
    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / (population_std_dev / math.sqrt(sample_size))

    # Calculate critical z-scores for the two-tailed test
    z_critical = norm.ppf(1 - significance_level / 2)  # e.g., for alpha=0.05, critical z â‰ˆ Â±1.96

    # Calculate the p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    return z_score, p_value, z_critical

# Input values
sample_mean = 105
population_mean = 100
population_std_dev = 15
sample_size = 30
significance_level = 0.05  # 5% significance level

# Perform the Z-test
z_score, p_value, z_critical = two_tailed_z_test(sample_mean, population_mean, population_std_dev, sample_size, significance_level)

# Display results
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")
print(f"Critical Z-Scores: Â±{z_critical:.2f}")

# Interpret results
if abs(z_score) > z_critical:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample mean and the population mean.")

# Visualization of the decision region
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)  # Standard normal distribution

# Plot the standard normal curve
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Standard Normal Curve', color='blue')

# Shade the rejection regions
plt.fill_between(x, 0, y, where=(x < -z_critical) | (x > z_critical), color='red', alpha=0.5, label='Rejection Region')

# Add labels and legend
plt.axvline(-z_critical, color='black', linestyle='--', label=f'Critical Z = -{z_critical:.2f}')
plt.axvline(z_critical, color='black', linestyle='--', label=f'Critical Z = +{z_critical:.2f}')
plt.axvline(z_score, color='green', linestyle='-', label=f'Z-Score = {z_score:.2f}')
plt.title('Two-Tailed Z-Test Decision Region')
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu_null, mu_alt, sigma, alpha, n):
    """
    Visualizes Type 1 and Type 2 errors for hypothesis testing.

    Parameters:
    - mu_null: Mean under the null hypothesis (H0).
    - mu_alt: Mean under the alternative hypothesis (H1).
    - sigma: Standard deviation of the population.
    - alpha: Significance level (Type 1 error rate).
    - n: Sample size.
    """
    # Calculate critical value (z-alpha for one-tailed test)
    z_alpha = norm.ppf(1 - alpha)
    critical_value = mu_null + z_alpha * (sigma / np.sqrt(n))

    # Calculate the probability of a Type 2 error (beta)
    beta = norm.cdf(critical_value, loc=mu_alt, scale=sigma / np.sqrt(n))

    # Generate x values for the distributions
    x = np.linspace(mu_null - 4 * sigma, mu_alt + 4 * sigma, 1000)

    # Null and alternative distributions
    null_dist = norm.pdf(x, mu_null, sigma / np.sqrt(n))
    alt_dist = norm.pdf(x, mu_alt, sigma / np.sqrt(n))

    # Plot the distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, null_dist, label='Null Hypothesis (H0)', color='blue')
    plt.plot(x, alt_dist, label='Alternative Hypothesis (H1)', color='red')

    # Highlight the rejection region for Type 1 error
    plt.fill_between(x, 0, null_dist, where=(x >= critical_value), color='blue', alpha=0.2, label='Type 1 Error (Î±)')

    # Highlight the acceptance region for Type 2 error
    plt.fill_between(x, 0, alt_dist, where=(x <= critical_value), color='red', alpha=0.2, label='Type 2 Error (Î²)')

    # Critical value
    plt.axvline(critical_value, color='black', linestyle='--', label=f'Critical Value = {critical_value:.2f}')

    # Add labels and legend
    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.xlabel('Test Statistic')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid()
    plt.show()

    print(f"Type 1 Error (Î±): {alpha}")
    print(f"Type 2 Error (Î²): {beta:.4f}")

# Example usage
# Parameters: mu_null, mu_alt, sigma, alpha, n
visualize_errors(mu_null=0, mu_alt=1, sigma=1, alpha=0.05, n=30)
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
        data (list or numpy array): The sample data.
        confidence (float): The confidence level (default is 0.95 for 95% confidence).

    Returns:
        tuple: (lower_bound, upper_bound) of the confidence interval.
    """
    sample_mean = np.mean(data)
    sample_size = len(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for unbiased standard deviation
    margin_of_error = stats.t.ppf((1 + confidence) / 2, df=sample_size-1) * (sample_std / np.sqrt(sample_size))

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
sample_data = [50, 52, 55, 60, 48, 47, 53, 58, 51, 49]
ci = confidence_interval(sample_data)
print(f"95% Confidence Interval: {ci}")
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy.stats import norm, t

def margin_of_error(data, confidence=0.95, known_variance=False):
    """
    Calculate the margin of error for a sample mean.

    Parameters:
        data (list or numpy array): The sample data.
        confidence (float): The confidence level (default is 0.95 for 95% confidence).
        known_variance (bool): If True, assumes population variance is known and uses Z-distribution.
                               Otherwise, uses T-distribution for small sample sizes.

    Returns:
        float: The margin of error.
    """
    sample_mean = np.mean(data)
    sample_size = len(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (unbiased)

    if known_variance:  # Z-distribution
        critical_value = norm.ppf((1 + confidence) / 2)
    else:  # T-distribution
        critical_value = t.ppf((1 + confidence) / 2, df=sample_size - 1)

    standard_error = sample_std / np.sqrt(sample_size)  # Standard error of the mean
    margin_error = critical_value * standard_error

    return margin_error

# Example usage
sample_data = [75, 80, 85, 90, 95, 100, 105]
confidence_level = 0.95  # 95% confidence level
moe = margin_of_error(sample_data, confidence=confidence_level, known_variance=False)

print(f"Margin of Error (95% confidence level): {moe}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior, likelihood, evidence):
    """
    Calculates the posterior probability using Bayes' Theorem.

    Parameters:
        prior (float): Prior probability of the hypothesis (P(H)).
        likelihood (float): Probability of the evidence given the hypothesis (P(E|H)).
        evidence (float): Probability of the evidence (P(E)).

    Returns:
        float: Posterior probability (P(H|E)).
    """
    posterior = (likelihood * prior) / evidence
    return posterior

# Example data for Bayesian inference
prior = 0.1  # Initial belief about the hypothesis (P(H))
likelihood = 0.8  # Likelihood of observing the evidence given the hypothesis (P(E|H))
evidence = 0.5  # Probability of observing the evidence (P(E))

posterior = bayes_theorem(prior, likelihood, evidence)
print(f"Posterior Probability (P(H|E)): {posterior:.2f}")
```

```python
#16. Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare the variances of two samples.

    Parameters:
        sample1 (array-like): Data for the first sample.
        sample2 (array-like): Data for the second sample.
        alpha (float): Significance level (default is 0.05).

    Returns:
        dict: F-statistic, critical value, and p-value.
    """
    # Calculate the sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Determine F-statistic
    if var1 > var2:
        f_stat = var1 / var2
        dfn, dfd = len(sample1) - 1, len(sample2) - 1  # Degrees of freedom
    else:
        f_stat = var2 / var1
        dfn, dfd = len(sample2) - 1, len(sample1) - 1

    # Calculate the critical value and p-value
    critical_value = f.ppf(1 - alpha, dfn, dfd)
    p_value = 1 - f.cdf(f_stat, dfn, dfd)

    return {
        "F-statistic": f_stat,
        "Critical Value": critical_value,
        "P-value": p_value
    }

# Example usage with random samples
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(loc=20, scale=5, size=30)  # Mean=20, Std=5, Size=30
sample2 = np.random.normal(loc=22, scale=8, size=30)  # Mean=22, Std=8, Size=30

# Perform the F-test
results = f_test(sample1, sample2)
print("F-test Results:")
print(f"F-statistic: {results['F-statistic']}")
print(f"Critical Value: {results['Critical Value']}")
print(f"P-value: {results['P-value']}")

# Interpretation
if results["F-statistic"] > results["Critical Value"]:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

import numpy as np
from scipy.stats import shapiro, levene
from statsmodels.stats.stattools import durbin_watson

def check_anova_assumptions(data, group_labels):
    """
    Checks the assumptions for ANOVA: normality, independence, and equal variance.

    Parameters:
        data (list): The combined list of all observations across groups.
        group_labels (list): A list of group labels corresponding to the data.

    Returns:
        None: Prints the results of the tests.
    """
    # Group the data by labels
    unique_groups = list(set(group_labels))
    grouped_data = [np.array([data[i] for i in range(len(data)) if group_labels[i] == group]) for group in unique_groups]

    print("Checking Normality (Shapiro-Wilk Test):")
    for i, group in enumerate(grouped_data):
        stat, p_value = shapiro(group)
        print(f"  Group {unique_groups[i]}: W-statistic = {stat:.4f}, p-value = {p_value:.4f}")
        if p_value < 0.05:
            print(f"  -> Group {unique_groups[i]} does NOT satisfy normality assumption (p < 0.05).")
        else:
            print(f"  -> Group {unique_groups[i]} satisfies normality assumption (p >= 0.05).")

    print("\nChecking Equal Variance (Levene's Test):")
    stat, p_value = levene(*grouped_data)
    print(f"  Levene's Test Statistic = {stat:.4f}, p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("  -> The groups do NOT have equal variances (p < 0.05).")
    else:
        print("  -> The groups have equal variances (p >= 0.05).")

    print("\nChecking Independence (Durbin-Watson Test - for residuals in regression):")
    dw_stat = durbin_watson(data)
    print(f"  Durbin-Watson Statistic = {dw_stat:.4f}")
    if dw_stat < 1.5 or dw_stat > 2.5:
        print("  -> Independence assumption may be violated (DW not within 1.5-2.5 range).")
    else:
        print("  -> Data satisfies independence assumption (DW within 1.5-2.5 range).")

# Example usage
data = [12, 15, 14, 10, 13, 20, 21, 19, 22, 23, 30, 28, 29, 31, 27]
group_labels = ['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'C', 'C', 'C', 'C', 'C']

check_anova_assumptions(data, group_labels)
```

```python
#20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
# results

import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Example dataset
data = {
    'Factor_A': ['Low', 'Low', 'Low', 'High', 'High', 'High', 'Low', 'Low', 'Low', 'High', 'High', 'High'],
    'Factor_B': ['A', 'A', 'B', 'B', 'A', 'A', 'A', 'B', 'B', 'A', 'A', 'B'],
    'Response': [20, 22, 19, 25, 30, 32, 18, 23, 21, 27, 29, 33]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Perform two-way ANOVA
model = ols('Response ~ C(Factor_A) + C(Factor_B) + C(Factor_A):C(Factor_B)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print the ANOVA table
print("ANOVA Results:")
print(anova_table)

# Visualization of interaction
sns.pointplot(x='Factor_A', y='Response', hue='Factor_B', data=df, dodge=True, markers=['o', 's'], linestyles=['-', '--'])
plt.title("Interaction Between Factor A and Factor B")
plt.xlabel("Factor A")
plt.ylabel("Mean Response")
plt.grid(True)
plt.show()
```

```python
#21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

def plot_f_distribution(dfn, dfd):
    """
    Plot the F-distribution with given degrees of freedom.

    Parameters:
        dfn (int): Degrees of freedom for the numerator.
        dfd (int): Degrees of freedom for the denominator.
    """
    x = np.linspace(0, 5, 1000)  # Range of x-values
    y = f.pdf(x, dfn, dfd)  # Compute F-distribution PDF

    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f"F-distribution (dfn={dfn}, dfd={dfd})", color='blue')
    plt.title("F-distribution")
    plt.xlabel("Value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
dfn, dfd = 5, 10  # Degrees of freedom for the numerator and denominator
plot_f_distribution(dfn, dfd)
```

```python
#22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.

import numpy as np
from scipy.stats import f_oneway
import matplotlib.pyplot as plt

# Sample data (Replace with your own data)
group1 = [12, 15, 14, 10, 13]
group2 = [20, 21, 19, 22, 23]
group3 = [30, 28, 29, 31, 27]

# Perform one-way ANOVA
f_statistic, p_value = f_oneway(group1, group2, group3)

# Print ANOVA results
print("One-way ANOVA Results:")
print("F-statistic:", f_statistic)
print("P-value:", p_value)

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("We reject the null hypothesis. There is a significant difference between the group means.")
else:
    print("We fail to reject the null hypothesis. There is no significant difference between the group means.")

# Visualization using boxplots
data = [group1, group2, group3]
labels = ['Group 1', 'Group 2', 'Group 3']

plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=labels, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Comparison of Group Means (Boxplot)")
plt.xlabel("Groups")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
#24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

import numpy as np
from scipy.stats import chi2

def variance_test(sample, hypothesized_variance, alpha=0.05):
    """
    Perform a hypothesis test for population variance using the Chi-square distribution.

    Parameters:
        sample (list or array): The sample data.
        hypothesized_variance (float): The hypothesized population variance.
        alpha (float): Significance level (default is 0.05).

    Returns:
        None: Prints the test results.
    """
    # Sample statistics
    sample_size = len(sample)
    sample_variance = np.var(sample, ddof=1)  # Unbiased estimate of variance

    # Chi-square test statistic
    chi_square_stat = (sample_size - 1) * sample_variance / hypothesized_variance

    # Degrees of freedom
    df = sample_size - 1

    # Critical values
    chi_critical_low = chi2.ppf(alpha / 2, df)  # Lower critical value
    chi_critical_high = chi2.ppf(1 - alpha / 2, df)  # Upper critical value

    # Interpretation
    print("Hypothesis Test for Population Variance:")
    print(f"Sample Variance: {sample_variance:.4f}")
    print(f"Chi-square Statistic: {chi_square_stat:.4f}")
    print(f"Critical Values: ({chi_critical_low:.4f}, {chi_critical_high:.4f})")

    if chi_square_stat < chi_critical_low or chi_square_stat > chi_critical_high:
        print("We reject the null hypothesis. The population variance is significantly different from the hypothesized variance.")
    else:
        print("We fail to reject the null hypothesis. The population variance is not significantly different from the hypothesized variance.")

# Example usage
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=50, scale=5, size=30)  # Mean=50, Std=5, Size=30
hypothesized_variance = 25  # Example hypothesized variance (standard deviation squared)

variance_test(sample_data, hypothesized_variance)
```

```python
#25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

import numpy as np
from statsmodels.stats.proportion import proportions_ztest

def z_test_proportions(successes, samples, alpha=0.05):
    """
    Perform a Z-test to compare proportions between two groups.

    Parameters:
        successes (list): The number of successes (e.g., counts of a specific outcome) for each group.
        samples (list): The total number of samples for each group.
        alpha (float): The significance level (default is 0.05).

    Returns:
        None: Prints the test results.
    """
    # Perform Z-test for proportions
    z_stat, p_value = proportions_ztest(successes, samples)

    # Print results
    print("Z-test for Proportions Results:")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpret results
    if p_value < alpha:
        print("We reject the null hypothesis. There is a significant difference between the proportions.")
    else:
        print("We fail to reject the null hypothesis. There is no significant difference between the proportions.")

# Example data
successes = [40, 30]  # Number of successes for each group
samples = [100, 120]  # Total number of samples for each group

# Perform the Z-test
z_test_proportions(successes, samples)
```

```python
#26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt

def f_test_and_visualization(data1, data2, alpha=0.05):
    """
    Perform an F-test for comparing variances of two datasets and visualize the results.

    Parameters:
        data1 (list or array): First dataset.
        data2 (list or array): Second dataset.
        alpha (float): Significance level (default is 0.05).

    Returns:
        None: Prints the results and shows a plot.
    """
    # Calculate variances and degrees of freedom
    var1, var2 = np.var(data1, ddof=1), np.var(data2, ddof=1)
    dfn, dfd = len(data1) - 1, len(data2) - 1  # Degrees of freedom

    # Determine F-statistic
    if var1 > var2:
        f_stat = var1 / var2
    else:
        f_stat = var2 / var1

    # Compute critical value and p-value
    critical_value = f.ppf(1 - alpha, dfn, dfd)
    p_value = 1 - f.cdf(f_stat, dfn, dfd)

    # Print results
    print("F-test Results:")
    print(f"Variance 1: {var1:.4f}, Variance 2: {var2:.4f}")
    print(f"F-statistic: {f_stat:.4f}")
    print(f"Critical Value (alpha={alpha}): {critical_value:.4f}")
    print(f"P-value: {p_value:.4f}")
    if f_stat > critical_value:
        print("Reject the null hypothesis: Variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: Variances are not significantly different.")

    # Visualization
    labels = ['Data 1', 'Data 2']
    variances = [var1, var2]

    plt.bar(labels, variances, color=['lightblue', 'pink'], edgecolor='black')
    plt.title('Comparison of Variances')
    plt.ylabel('Variance')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Example data
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=5, size=30)  # Mean=50, Std=5, Size=30
data2 = np.random.normal(loc=55, scale=8, size=30)  # Mean=55, Std=8, Size=30

# Perform F-test and visualize results
f_test_and_visualization(data1, data2)
```

```python
#27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

import numpy as np
from scipy.stats import chisquare

# Simulate observed data (e.g., counts of different categories)
np.random.seed(42)  # For reproducibility
observed = np.random.choice([10, 20, 30, 40], size=100, p=[0.1, 0.4, 0.4, 0.1])
unique, counts = np.unique(observed, return_counts=True)  # Count occurrences
observed_counts = counts

# Define the expected frequencies (uniform or hypothesized distribution)
total = np.sum(observed_counts)
expected_counts = [total * prob for prob in [0.1, 0.4, 0.4, 0.1]]

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed_counts, f_exp=expected_counts)

# Print results
print("Chi-square Test for Goodness of Fit Results:")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("We reject the null hypothesis. The observed data does not fit the expected distribution.")
else:
    print("We fail to reject the null hypothesis. The observed data fits the expected distribution.")
```
